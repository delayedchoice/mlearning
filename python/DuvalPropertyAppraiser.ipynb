```python
import os
import tarfile
from six.moves import urllib



BASE_HOUSE_URL = "http://apps.coj.net/PAO_PropertySearch/Basic/Detail.aspx?RE="
HOUSING_FILE = os.path.join("datasets", "duval", "data-0418.2018.csv")
HOUSING_PATH = os.path.join("datasets", "duval", "downloads")
    
def fetch_house_data(house_id,housing_url=BASE_HOUSE_URL):
    house_id=house_id.replace("-", "")
    if not os.path.isdir(HOUSING_PATH):
        os.makedirs(HOUSING_PATH)
    target_file = os.path.join(HOUSING_PATH, house_id)
    urllib.request.urlretrieve(housing_url + house_id, target_file)
```

```python
import pandas as pd

def load_housing_data(housing_path=HOUSING_FILE):
    return pd.read_csv(housing_path)
```

```python
housing["income_cat"] = np.ceil(housing["median_income"] / 1.5)
housing["income_cat"].where(housing["income_cat"] < 5, 5.0, inplace=True)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True)
```

```python
housing = strat_train_set.copy()
```
